```python
def addoptions():
	
	import optparse
	#import glob
	
	#Initialize command line parser
	parser = optparse.OptionParser()
	
	#Adding the options
	parser.add_option('-i', type="string", help='input data file for PCA calculation' )
	parser.add_option('-r', type="string", help='If row header present, then pass 1 else 0' )
	parser.add_option('-c', type="string", help='If column header present, then pass 1 else 0' )
	args, remainder = parser.parse_args()
	
	#Assigning the command line options to variables
	fname = args.i
	rowheader = args.r
	colheader = args.c
	
	return [fname, rowheader, colheader]
```

```python
def getdata(fname, rowheader, colheader):
	
	#read the data file in variable lines
	inf = open(fname, 'r')
	lines = inf.readlines()
	inf.close();


	newlines = [];

	#if header is present
	if rowheader == '1':
		#Get the list of variable names from the header (lines[0]).
		#Skip the first element for column label
		Variables = lines[0].strip().split()[1:]

		newlines = lines[1:]

	else:
		#if header is not present
		newlines = lines;

	#Read file line by line, skip first line
	for l in newlines:

		#Split the line using tabs/spaces/multiple spaces as delimeter
		toks = l.strip().split();

		if colheader=='1':
			#Map the split to the folating points data type for mathematical operations
			temppoints = list(map(float,toks[1:]))
		else:
			#Map the split to the folating points data type for mathematical operations
			temppoints = list(map(float,toks))

		#Append the data points in the global list
		Datapoints.append(temppoints);

	#return the global list
	return Datapoints, Variables
```
